```python
def standardize(x,override = False):
    if not override:
        assert(x.shape[0]<x.shape[1])

    dem = np.std(x,0)
    if (dem == 0).any():
        dem = np.where(np.std(x, 0) == 0, 1, np.std(x, 0))
    return (x - np.mean(x, 0))/dem
```

# Get bootstrapped coefs

```python
final_res_dict = res_dict['coef_bootstrap']
folds = list(final_res_dict.keys())

if len(folds)>0:
    coef_arr = np.concatenate([list(final_res_dict[fold]['coef'].values()) for fold in folds],1)
    medians = np.median(coef_arr, 1)
    median = pd.Series(np.median(medians,0), index = coef_names)
    mad = pd.Series(get_mad(medians,0), index = coef_names)
    upper, lower = get_median_ci(median, mad, 'coef')
else:
    final_res_dict = mb.nested_cv_func(model, x, y, name=name, dtype = 'metabolites', optim_param = 'auc', plot_lambdas=False, learn_var = 'C')
    coef_arr = final_res_dict['coef']
    median = pd.Series(np.median(coef_arr,0), index = coef_names)
    mad = pd.Series(get_mad(coef_arr,0), index = coef_names)
    upper, lower = get_median_ci(median, mad, 'coef')

ix_median = np.argsort(-np.abs(median))
ix_median_nzero = [ix for ix in ix_median if median[ix]!=0]

ix_MAD = np.argsort(-np.abs(mad))
ix_MAD_all = [ix for ix in ix_MAD if ix not in ix_median_nzero]
ix_tot = np.concatenate([ix_median_nzero, ix_MAD_all])

median_lo = np.round(np.exp(median),8)
mad_lo = np.round(np.exp(mad),8)
med_cdi_lo = list(zip(lower, upper))

fin_dict = {'Median OR':median_lo, 'MAD interval': med_cdi_lo}
coef_df = pd.DataFrame(fin_dict).iloc[ix_tot]
coef_df.to_csv('outputs/SigBiomarkers_' + str(datetime.now()).split(' ')[0] + '.csv')
```

```python
CIvec = []
Qvec = []
for metabolite in med_df.columns.values:
    col = med_df[metabolite]
    ix_sort = np.argsort(-np.abs(col))
    sort = np.array(col)[ix_sort]
    ix_upper = np.int(np.floor(len(sort)*.05))
    ix_lower = np.int(np.ceil(len(sort)*.95))
    CIvec.append((np.round(np.exp(sort[ix_lower]),8), np.round(np.exp(sort[ix_upper]),8)))
    
    ix_25 = np.int(np.floor(len(sort)*.75))
    ix_75 = np.int(np.ceil(len(sort)*.25))
    Qvec.append((np.round(np.exp(sort[ix_25]),8), np.round(np.exp(sort[ix_75]),8)))
    

coef_df['5%, 95% Interval'] = CIvec
coef_df['25%, 75% Quartiles'] = Qvec
coef_df.to_csv('outputs/SigBiomarkers_wIntervals' + str(datetime.now()).split(' ')[0] + '.csv')
coef_df.head()
```

```python
def get_median_ci(median, mad, param, z=1.96):
    if param == 'coef':
        upper = np.round(np.exp(median + mad),4)
        lower = np.round(np.exp(median - mad),4)
    else:
        upper = np.round(median + mad,4)
        lower  = np.round(median - mad,4)
    return upper, lower
```

```python
def get_ci(val,z=1.96):
    mean = np.mean(val)
    std = np.std(val)
    l = len(val)
    CI_top = np.round(mean + z*(std)/np.sqrt(l),4)
    CI_bot = np.round(mean - z*(std)/np.sqrt(l),4)
    return (CI_bot, CI_top)
```

```python
key = list(new_res_dict.keys())[0]

    # get medians across folds
med_dict = {}
stat_dict = {}
folds = np.arange(len(new_res_dict[key]))
for param in new_res_dict[key][0]:
    if param == 'model' or param == 'optimizer':
        continue
    stat_dict[param] = {}
    med_dict[param] = {}
    if isinstance(new_res_dict[key][0][param],list):
        medians = np.median(np.vstack([new_res_dict[key][fold][param] for fold in folds]),0)
        median = np.median(medians)
        mad = get_mad(medians)
        upper, lower = get_median_ci(median, mad, param)
    elif isinstance(new_res_dict[key][0][param],dict):
        arr = np.concatenate([list(new_res_dict[key][fold][param].values()) for fold in folds],1)
        medians = np.median(arr, 1)
        med_dict[param] = medians
        median = pd.Series(np.median(medians,0), index = coef_names)
        mad = pd.Series(get_mad(medians,0), index = coef_names)
        upper, lower = get_median_ci(median, mad, param)
    elif isinstance(new_res_dict[key][0][param],float):
        median = np.median([new_res_dict[key][fold][param] for fold in folds])
        mad = get_mad([new_res_dict[key][fold][param] for fold in folds])
        upper, lower = get_median_ci(median, mad, param)

    stat_dict[param]['Median'] = median
    stat_dict[param]['Mad'] = mad
    stat_dict[param]['lower_ci'] = lower
    stat_dict[param]['upper_ci'] = upper 

```
